```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

cwd = os.getcwd()  # Get the current working directory (cwd)
files = os.listdir(cwd)  # Get all the files in that directory
print("Files in %r: %s" % (cwd, files))
```

```python
f = open('./duy/california_cities.csv')
cities = pd.read_csv(f)
cities.head()
cities['area_total_km2'].sort_values(ascending=False)
```

```python
# extract latd = vi do and longd = kinh do
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
lon[:10]
# Plot using pyplot API
#figure chinh khung
plt.figure(figsize=(8, 6))

plt.scatter(lon, lat,
            c=np.log10(population),
            cmap='magma',
            alpha=0.5,
            linewidth=0,
            s=area)

# cmap = dãy màu 1->6 type giống 7 sắc cầu vòng
# Lam cho dep hon
plt.axis('equal')
plt.xlabel('longitude')
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}$population')
plt.clim(4,6)
plt.title('California Cities: Population and Area Distribution')

#Create A legend for cities'sizes
area_rang =[50,200,300,500]
for area in  area_rang:
    plt.scatter([],[],
                label=str(area),
                s=area,
                alpha=0.5,
                c='k')
plt.legend(title='City Area',labelspacing=1.5)
```
